```python
# Drop variabel NaN
# missing value >30% akan dilakukan drop 
house_cleaned_afterdrop = house.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu'])

#Melihat persentase missing value setiap kolom setelah drop na pada missing value > 30%

persentase_house = house_cleaned_afterdrop.isna().sum()/len(house)*100
persentase_house

persentase_house_ascending = persentase_house.sort_values(ascending=False)
house_afterdrop = persentase_house_ascending.head(20)
house_afterdrop
```

```python
### Tulis kode Anda di sini
# Visualisasi histogram untuk variabel numerikal 'SalePrice'
plt.figure(figsize=(10, 6))
sns.histplot(house_cleaned_afterdrop['SalePrice'], bins=10, kde=True)
plt.title('Distribusi Harga Jual Properti')
plt.xlabel('Harga Jual($)')
plt.ylabel('Frekuensi')
plt.show()

# Analisis sederhana
# Median
median_sale_price = house_cleaned_afterdrop['SalePrice'].median()
# Mean
mean_sale_price = house_cleaned_afterdrop['SalePrice'].mean()
# Std Dev
std_sale_price = house_cleaned_afterdrop['SalePrice'].std()

print("Harga jual median:", median_sale_price)
print("Harga jual rata-rata (mean):", mean_sale_price)
print("Standar deviasi dari harga jual properti:", std_sale_price)
```

#### **Analisis tentang variabel `SalePrice`**

* Dari data yang diberikan, kita memiliki beberapa statistik deskriptif tentang variabel SalePrice, yang merupakan variabel target yang ingin diprediksi dalam suatu model. Variabel SalePrice mewakili harga properti yang dijual, yang merupakan informasi krusial dalam industri real estate.

Median Harga Jual: Median harga jual adalah 163,000.0. Median adalah nilai tengah dari distribusi harga jual properti yang diurutkan dari yang terendah hingga tertinggi. Median seringkali lebih representatif daripada mean (rata-rata) dalam kasus di mana distribusi harga tidak simetris atau memiliki pencilan yang signifikan.
Rata-Rata Harga Jual: Rata-rata harga jual adalah sekitar 180,921.20. Rata-rata adalah jumlah dari semua nilai harga jual properti yang dibagi dengan jumlah total properti. Rata-rata memberikan gambaran tentang tendensi sentral dari harga properti, tetapi rentan terhadap pengaruh outlier atau nilai ekstrem.
Standar Deviasi Harga Jual: Standar deviasi adalah sekitar 79,442.50. Standar deviasi mengukur seberapa jauh sebaran harga jual properti dari rata-rata. Semakin tinggi standar deviasi, semakin besar variabilitas harga jual properti dari nilai rata-rata. Standar deviasi yang tinggi menunjukkan bahwa harga properti dapat bervariasi secara signifikan dari satu properti ke properti lainnya.
Dengan informasi ini, dapat disimpulkan bahwa variabel SalePrice memiliki distribusi yang cenderung agak bervariasi dengan standar deviasi yang cukup tinggi. Hal ini menunjukkan bahwa dalam pengembangan model prediksi harga properti, perlu mempertimbangkan pengelompokan atau klasifikasi yang tepat untuk memahami pola-pola harga yang mungkin terjadi. Selain itu, penggunaan median sebagai ukuran tendensi sentral bisa lebih stabil daripada rata-rata dalam hal mendapatkan gambaran yang representatif tentang harga properti, terutama dalam kasus-kasus di mana ada nilai ekstrem atau perbedaan besar dalam harga properti.

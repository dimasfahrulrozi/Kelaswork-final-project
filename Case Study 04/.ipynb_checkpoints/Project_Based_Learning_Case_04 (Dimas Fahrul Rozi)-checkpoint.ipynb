```python
# Importing the libraries, fill the import name libraries below
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import SGD

#Import keras libraries that you need
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU, Bidirectional
import math
from sklearn.metrics import mean_squared_error

```

# EDA

```python
# Some functions to help out with this code
def plot_predictions(test,predicted):
    plt.plot(test, color='red',label='Real AABA Stock Price')
    plt.plot(predicted, color='blue',label='Predicted AABA Stock Price')
    plt.title('AABA Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('AABA Stock Price')
    plt.legend()
    plt.show()
    #fill the function to predict
def create_LSTM_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def predict_stock_prices_LSTM(train_data, test_data):
    # Assuming train_data and test_data are numpy arrays or pandas dataframes
    X_train, y_train = train_data[:, :-1], train_data[:, -1]
    X_test, y_test = test_data[:, :-1], test_data[:, -1]

    # Reshape data for LSTM input (samples, time steps, features)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Create and fit LSTM model
    model = create_LSTM_model(input_shape=(X_train.shape[1], 1))
    model.fit(X_train, y_train, epochs=100, batch_size=32)

    # Make predictions on test data
    predicted_prices = model.predict(X_test)

    return predicted_prices.flatten()

# Example usage
# Assuming you have train_data and test_data prepared
# train_data and test_data should be numpy arrays or pandas dataframes with the same number of features
# Call the predict_stock_prices_LSTM function to get predicted prices
# Then pass the test_data and predicted_prices to plot_predictions
# plot_predictions(test_data[:, -1], predicted_prices)

def rmse_eval(test,predicted):
    rmse = math.sqrt(mean_squared_error(test, predicted))
    print("The root mean squared error is {}.".format(rmse))
    return rmse
    #fill the function for evaluation model


# Example usage
# Assuming you have test_data and predicted_prices from your prediction function
# rmse = rmse_eval(test_data[:, -1], predicted_prices)
   
```

```python
# Import Dataset
dataset = pd.read_csv('AABA_2006-01-01_to_2018-01-01.csv', index_col ='Date', parse_dates = True) #import dataset based on your directory
dataset.head()
```

```python
# Checking for missing values, choose for data training and set in kind of year
#a : until year training 
#b : year for testing
training_set = dataset[:'2016'].iloc[:,1:2].values
test_set = dataset['2017':].iloc[:,1:2].values
```

```python
#Checking for stock's market company with line plot
# We have chosen 'High' attribute for prices. Let's see what it looks like
dataset["High"][:'2016'].plot(figsize=(16,4),legend=True)
dataset["High"]['2017':].plot(figsize=(16,4),legend=True)
plt.legend(['Training set (Before 2017)','Test set (2017 and beyond)'])
plt.title('AABA stock price')
plt.show()
```

```python
# Scaling with minmaxscaller for data train
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
#checking shape of data train
training_set_scaled.shape
```

```python
# Initialize empty lists for X_train and y_train
X_train = []
y_train = []

# Determine the maximum index based on the length of training_set_scaled
max_index = len(training_set_scaled) - 60  # Subtract 60 to avoid going out of bounds

for i in range(60, max_index):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])

# Convert lists to arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
# Reshaping X_train before modelling
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
```

## Modelling

```python
# Sequential Modelling
model = Sequential()
# First LSTM layer with Dropout regularisation
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))
# Second LSTM layer
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
# Third LSTM layer, fill the code below
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
# Fourth LSTM layer, fill the code below
model.add(LSTM(units=50))
model.add(Dropout(0.2))
# The output layer
model.add(Dense(units=1))
# Compiling
model.compile(optimizer='rmsprop',loss='mean_squared_error')
# Train fitting for the model
model.fit(X_train,y_train,epochs=50,batch_size=32)
```

```python
# Prepare for data test similar way for data train
dataset_total = pd.concat((dataset["High"][:'2016'],dataset["High"]['2017':]),axis=0)
# Extract the test data from dataset_total
test = dataset_total['2017':]
inputs = dataset_total[len(dataset_total)-len(test_set) - 60:].values
inputs = inputs.reshape(-1,1)
#Transform the inputs
inputs  = sc.transform(inputs)
```

```python
#checking data test shape
inputs.shape
```

### Prediction

```python
# Preparing X_test and predicting the prices of the stock's that you choose
X_test = []
for i in range(60,311):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
#Predicting stocks price
predicted_stock_price = model.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
# Visualizing the results for prediction 
visualize = plot_predictions(test_set,predicted_stock_price)
```

```python
# Evaluating our model with RMSE function above
rmse_eval(test_set,predicted_stock_price)
# Evaluating our model
```

Insight:
Interpretasi Nilai RMSE:
Dalam konteks ini, nilai RMSE sebesar 4.35 menunjukkan tingkat kesalahan prediksi model sebesar 4.35 unit pada skala yang sama dengan variabel yang diprediksi.

Performa Model Berdasarkan RMSE:
Semakin rendah nilai RMSE, semakin baik performa model dalam memprediksi nilai sebenarnya dari data.

Advice: 
1) Evaluasi secara kritis: untuk mengevaluasi nilai RMSE dengan konteks yang tepat dan kritis. Misalnya, nilai RMSE sebesar 4.35 mungkin menunjukkan tingkat kesalahan yang dapat diterima dalam beberapa kasus, tetapi mungkin tidak dalam kasus lain. jika tidak kita dapat melakukan: Optimasi model atau uji coba variasi

2) Optimisasi model: Jika nilai RMSE masih terlalu tinggi maka, dapat dipertimbangkan untuk melakukan optimisasi lebih lanjut pada model Anda. Ini bisa melibatkan peningkatan fitur, pemilihan model yang lebih canggih, atau penyesuaian parameter model yang lebih baik.

3) Uji coba variasi: Lakukan uji coba dengan variasi model dan metode evaluasi lainnya untuk mendapatkan gambaran yang lebih lengkap tentang kinerja model. Selain RMSE, kita juga dapat mempertimbangkan metrik evaluasi lainnya seperti R-squared, MAE (Mean Absolute Error), atau metrik lain yang sesuai dengan kebutuhan analisis Anda.
